```python
import csv
import numpy as np
from seaborn import lineplot as lp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def workData(p,q):
    data = []
    with open("2_combination.csv") as h:
        reader = csv.reader(h)
        for line in reader:
            if float(line[1])==p and float(line[2])==q:
                data.append(line)

    df = pd.DataFrame(data, columns= ['SimNum','P','Q','t',"I"])

    for i in range(50):
        plotData = [float(ele) for ele in df.loc[df["SimNum"]==str(i)]["I"].tolist()]
        lp(plotData)
```

```python
P = np.linspace(0,.8,101)
Q = np.linspace(.8,.999,101)

print(P)
print("----------------")
print(Q)
```
