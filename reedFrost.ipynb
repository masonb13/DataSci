```python
import numpy as np
import pandas as pd
from scipy.stats import binom
import matplotlib.pyplot as plt
import statistics as stat
```

```python
#new infections = binom(S,P) S-number of susceptibles, P-probability of infection
```

```python
def reedFrost(i, s, q, numRep, numGen):
    r = 0

    I = [[i] for x in range(numRep)]
    S = [[s] for x in range(numRep)]
    R = [[r] for x in range(numRep)]

    for y in range(numRep):
        for z in range(numGen):
            I[y].append(int(binom.rvs(S[y][-1],(1-q**(I[y][-1])), size=1)))
            S[y].append(S[y][-1]-I[y][-1])
            R[y].append(R[y][-1]+I[y][-2])
    print(len(S[y]))
    print(len(I[y]))
    return [I,S,R]
```

```python
#Variables
I = [1,2,3,4,5]
S = [500, 1000,2000,3000,4000]
Q = [.2,.4,.6,.8,.9999]


def varyVar(I,S,Q,numRep, numGen, i, s, q):
    result = [[] for i in range(3)]
    output = [[] for i in range(3)]
    for num in I:
        ans = reedFrost(num,s,q, numRep, numGen)[-1]
        output[0].append(sum(ans))
        result[0].append(ans)
    for num in S:
        ans = reedFrost(i,num,q, numRep, numGen)[-1]
        output[1].append(sum(ans))
        result[1].append(ans)
    for num in Q:
        ans = reedFrost(i,s,num, numRep, numGen)[-1]
        output[2].append(sum(ans))
        result[2].append(ans)

    return output



result = varyVar(I, S, Q, numRep, numGen, i, s, q)
print(result)

# x = [i for i in range(numGen+1)]
# plt.hist(result[0])
# plt.text(I)

# plt.show()

```
